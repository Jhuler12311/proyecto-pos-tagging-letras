# Comparación Completa POS Tagging: spaCy vs NLTK

Este notebook incluye:
- Carga de resultados
- Comparación cuantitativa
- Visualizaciones
- Conclusiones técnicas


```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar tabla comparativa
df_compare = pd.read_csv('../data/processed/comparacion_etiquetas.csv', index_col=0)
df_compare = df_compare[df_compare.index != '']
df_compare.head()
```

## Gráfico 1: Distribución de etiquetas spaCy


```python
plt.figure()
df_compare['spaCy_Universal'].sort_values(ascending=False).head(15).plot(kind='bar')
plt.title('Top 15 Etiquetas spaCy (Universal POS)')
plt.ylabel('Frecuencia')
plt.show()
```

## Gráfico 2: Distribución de etiquetas NLTK


```python
plt.figure()
df_compare['NLTK_PennTreebank'].sort_values(ascending=False).head(15).plot(kind='bar')
plt.title('Top 15 Etiquetas NLTK (Penn Treebank)')
plt.ylabel('Frecuencia')
plt.show()
```

## Gráfico 3: Comparación Directa


```python
plt.figure()
df_compare[['spaCy_Universal', 'NLTK_PennTreebank']].sum().plot(kind='bar')
plt.title('Total Etiquetas Detectadas por Modelo')
plt.ylabel('Total')
plt.show()
```

## Análisis Técnico

Aunque se implementó detección dinámica de idioma en spaCy (español e inglés),
la comparación sigue mostrando diferencias estructurales importantes.

Razones:
1. NLTK utiliza un modelo entrenado principalmente en inglés.
2. Cuando procesa letras en español, la asignación de etiquetas es limitada o imprecisa.
3. spaCy cambia correctamente entre modelos multilingües.

Esto demuestra que la diferencia no es un error del pipeline,
sino una consecuencia del entrenamiento y arquitectura de cada herramienta.


## Conclusión Final

spaCy demuestra:
- Mayor adaptabilidad multilingüe
- Mejor distribución morfológica
- Mayor coherencia en corpus mixtos

NLTK presenta limitaciones en escenarios multilingües.

Para proyectos de análisis de letras bilingües,
spaCy es la herramienta técnicamente más robusta.
