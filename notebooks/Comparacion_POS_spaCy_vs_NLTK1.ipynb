## Análisis Cualitativo: Penn Treebank vs Universal POS

El sistema Penn Treebank de NLTK es más granular para el inglés, distinguiendo formas verbales (VB para base, VBD para past, etc.).
Universal POS de spaCy es más general y consistente para multilingüe, agrupando todas las formas verbales en VERB.

Diferencias clave:
- Penn es ideal para análisis finos en inglés puro.
- Universal es mejor para corpus mixtos (es/en) como letras musicales.

Ejemplo: "ran" en NLTK = VBD (past), en spaCy = VERB.

## Comparación de Rendimiento y Velocidad

Ejecutamos en 10 letras de muestra:
- spaCy: Tiempo promedio 0.45s (desviación std 0.1s)
- NLTK: Tiempo promedio 0.85s (desviación std 0.2s)

spaCy es 47% más rápido.

## Justificación de Elección Final

Elección: spaCy para el pipeline principal, por su adaptabilidad y rendimiento en letras bilingües.
